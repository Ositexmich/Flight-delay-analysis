```python

import re

# 1. Read all tables
url    = "https://en.wikipedia.org/wiki/List_of_the_busiest_airports_in_the_United_States"
tables = pd.read_html(url, header=0)

# 2. Identify large- and medium-hub tables by inspecting their columns
#    (you already saw these are tables[1] and tables[2])
large_df  = tables[1].copy()
medium_df = tables[2].copy()

# 3. Utility to find the column whose name starts with a given year (e.g. "2023")
def pick_year_col(df, year=2023):
    pattern = re.compile(rf'^{year}')
    for col in df.columns:
        if pattern.match(col):
            return col
    raise KeyError(f"No column begins with {year}")

# 4. Locate the “2023” column in each
year_col_large  = pick_year_col(large_df, 2023)
year_col_medium = pick_year_col(medium_df, 2023)

# 5. Utility to slice & clean one hub-class table
def prepare_hubs(df, hub_type, year_col):
    # detect rank, airport name, and IATA columns
    rank_col    = next(c for c in df.columns if "Rank" in c)
    airport_col = next(c for c in df.columns if c.lower().startswith("airports"))
    iata_col    = next(c for c in df.columns if "IATA" in c)
    
    sub = df[[rank_col, airport_col, iata_col, year_col]].copy()
    sub.columns = ["Rank", "Airport", "IATA", "Enplanements"]
    
    # strip commas and convert to int
    sub["Enplanements"] = (
        sub["Enplanements"]
        .astype(str)
        .str.replace(",", "")
        .astype(int)
    )
    
    sub["Hub_Type"] = hub_type
    return sub

# 6. Apply and concatenate
large_hubs  = prepare_hubs(large_df,  "Large",  year_col_large)
medium_hubs = prepare_hubs(medium_df, "Medium", year_col_medium)
hub_traffic = pd.concat([large_hubs, medium_hubs], ignore_index=True)

# 7. Inspect the combined DataFrame
print(hub_traffic.head(10))
```

### Hypothesis Testing

#### Height of airport elevation Vs  Arrival and Departure Delay

```python
from scipy.stats import pearsonr
import statsmodels.api as sm



# Drop NaNs
df_depart = merged_New[['airportfrom', 'elevation_ft', 'delay']].dropna()
df_arrive = merged_New[['airportto', 'elevation_ft', 'delay']].dropna()

# ----- DEPARTURES -----
print("\n📦 Departure Flights Analysis")

# Correlation test
corr_depart, pval_depart = pearsonr(df_depart['elevation_ft'], df_depart['delay'])
print(f"Pearson Correlation (Departures): {corr_depart:.4f}, p-value: {pval_depart:.4g}")

# Linear Regression
X_depart = sm.add_constant(df_depart['elevation_ft'])
model_depart = sm.OLS(df_depart['delay'], X_depart).fit()
print(model_depart.summary())

# ----- ARRIVALS -----
print("\n📦 Arrival Flights Analysis")

corr_arrive, pval_arrive = pearsonr(df_arrive['elevation_ft'], df_arrive['delay'])
print(f"Pearson Correlation (Arrivals): {corr_arrive:.4f}, p-value: {pval_arrive:.4g}")

X_arrive = sm.add_constant(df_arrive['elevation_ft'])
model_arrive = sm.OLS(df_arrive['delay'], X_arrive).fit()
print(model_arrive.summary())
```

With the P values greater than 0.05(p > 0.05),Indicates that the height of airport elevation has nothing to do with delay arrival and delay departure.

#### Number of Runways Vs Flight delay

```python

# STEP 1: Count number of runways per airport (departure)
runway_counts = merged_New.groupby('airportfrom').size().reset_index(name='runway_count')

# STEP 2: Get average delay per airport
avg_delay = merged_New.groupby('airportfrom')['delay'].mean().reset_index(name='avg_delay')

# STEP 3: Merge both
airport_data = pd.merge(runway_counts, avg_delay, on='airportfrom')

# STEP 4: Run correlation and regression
corr, pval = pearsonr(airport_data['runway_count'], airport_data['avg_delay'])
print(f"📊 Pearson Correlation: {corr:.4f}, p-value: {pval:.4g}")

# Regression
X = sm.add_constant(airport_data['runway_count'])
model = sm.OLS(airport_data['avg_delay'], X).fit()
print(model.summary())
```

From the above result with p value less than 0.05 (p < 0.05), we shall reject the null hypothesis that number of runways does not affect flight delays.

### Duration of Flight vs Flight delays

```python
# Drop rows with missing values
df = merged_New[['airportfrom', 'airportto', 'length', 'delay']].dropna()

# ---- 1. DEPARTURE PERSPECTIVE ----
from_group = df.groupby('airportfrom').agg(
    avg_length=('length', 'mean'),
    avg_delay=('delay', 'mean')
).reset_index()

# Correlation
corr_from, pval_from = pearsonr(from_group['avg_length'], from_group['avg_delay'])
print(f"📦 DEPARTURES: Pearson correlation = {corr_from:.4f}, p-value = {pval_from:.4g}")

# Regression
X_from = sm.add_constant(from_group['avg_length'])
model_from = sm.OLS(from_group['avg_delay'], X_from).fit()
print(model_from.summary())


# ---- 2. ARRIVAL PERSPECTIVE ----
to_group = df.groupby('airportto').agg(
    avg_length=('length', 'mean'),
    avg_delay=('delay', 'mean')
).reset_index()

# Correlation
corr_to, pval_to = pearsonr(to_group['avg_length'], to_group['avg_delay'])
print(f"📦 ARRIVALS: Pearson correlation = {corr_to:.4f}, p-value = {pval_to:.4g}")

# Regression
X_to = sm.add_constant(to_group['avg_length'])
model_to = sm.OLS(to_group['avg_delay'], X_to).fit()
print(model_to.summary())
```

From the p values from both the departure airport(p value < 0.05) and the arriving airport(p value> 0.05), We shall reject the Null Hypothesis on the departure airport but fail to reject the Null Hypotheis on the arriving airport. What this mean is the duration of flight does not affecting the departure airport, however, it does affecting the arriving airport. Perhaps due to weather during the travel time.

#### Handling overfitting of Decision Tree using Stratified 5-fold method

```python
from sklearn.model_selection import StratifiedKFold
from scipy.stats import mode

# Set up stratified 5-fold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Store predictions for the test set from each fold
test_preds = []

# Perform stratified 5-fold training
for fold, (train_idx, val_idx) in enumerate(skf.split(X_train, y_train)):
    print(f"Training fold {fold + 1}...")

    X_fold_train, X_fold_val = X_train.iloc[train_idx], X_train.iloc[val_idx]
    y_fold_train, y_fold_val = y_train.iloc[train_idx], y_train.iloc[val_idx]

    # Train the model
    model = base_model.fit(X_fold_train, y_fold_train)

    # Predict on test set
    y_test_pred = model.predict(X_test)
    test_preds.append(y_test_pred)

# Convert to array: shape (5, n_samples)
test_preds = np.array(test_preds)

# Majority vote across the 5 predictions (axis=0 means vote per test sample)
final_predictions = mode(test_preds, axis=0).mode

# Evaluate final prediction
print("\n🎯 Final Prediction (5-Fold Voting) — TEST")
print("Accuracy:", accuracy_score(y_test, final_predictions))
print(classification_report(y_test, final_predictions))
```
